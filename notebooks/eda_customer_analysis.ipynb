```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/customers.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

The average annual income is higher than the median, indicating a slightly right-skewed income distribution.

```python
plt.figure()
plt.hist(df["annual_income"], bins=5)
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.title("Distribution of Annual Income")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df["annual_income"])
plt.title("Boxplot of Annual Income")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="gender", y="spending_score", data=df)
plt.title("Spending Score by Gender")
plt.show()
```

```python
df.groupby("category")["spending_score"].mean().sort_values(ascending=False)
```

```python
plt.figure()
sns.heatmap(df[["age", "annual_income", "spending_score"]].corr(),
            annot=True,
            cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
